```python
import pandas
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.font_manager as fm
plt.style.use("style/hcilab.mplstyle")
mpl.rcParams['pdf.fonttype'] = 42
minionpro = fm.FontProperties(fname='./style/minion-pro/MinionPro-Regular.otf')
```

```python
data = pandas.read_csv("./plan_15sec.csv")
```

```python
def plot_data(ax,veropt,pddl,label):
    ax.set_facecolor('#f8f9f9')
    ax.set_ylabel(label,fontproperties=minionpro, size=15)
    means = [.25,1,1.75]
    veropt_idx = [m-.2 for m in means]
    pddlopt_idx = [m+.2 for m in means]
    
    # Add veropt results
    ax.bar(veropt_idx,veropt,color='#829abf')

    # Add pddl results
    ax.bar(pddlopt_idx,veropt,color='#dcdddd')

fig, axs = plt.subplots(4,1,figsize=(4,8),sharex=True)
compute_plot = axs[0]
time_plot = axs[1]
cost_plot = axs[2]
score_plot = axs[3]
xticks = ["Kitting","Assembly","Repair"]

# Compute Time
veropt = data.groupby(["planner","plan"]).mean().loc["veropt","compute time"]
pddlopt = data.groupby(["planner","plan"]).mean().loc["veropt","compute time"]
plot_data(compute_plot,veropt,pddlopt,"Compute Time")

# Plan Time
plot_data(time_plot,[],[],"Plan Time")

# Plan Cost
plot_data(cost_plot,[],[],"Plan Cost")

# Plan Score
plot_data(score_plot,[],[],"Plan Score")


plt.xticks([.25,1,1.75],xticks,fontproperties=minionpro, size=12)
plt.xlim(0,2)

plt.tight_layout()
```
